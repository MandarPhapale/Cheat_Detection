```python
#Code to create dummy logs (Messages have keywords)


import json
from datetime import datetime, timedelta
import random
import time

# Add header to output var
# [{"type":"Title","title":"Whiznook game log","timestamp":[18944,261479454]},{"type":"Message","original":{"author":"rtrt","message":"rtrtr","clientId":"1","timestamp":[18951,701787159],"messageDelay":0.02},"timestamp":[18951,701962624]}
# ,{"type":"Message","original":{"author":"rtrt","message":"reouhvrv","clientId":"1","timestamp":[18954,966706393],"messageDelay":0.03},"timestamp":[18954,966755544]}
msgP_array = []
logstart = datetime.now() - timedelta(hours=24)

for x in range(50):
	authors = ["James", "Mandar", "Harry", "Anastasia"]
	MsgsP = ["Try exploring the room picture", "I found a diagram with letters", "Do you see a map", "there is something currently invisible", "on the desk to the left of the keyboard there is a hotspot", "For the Hydraulics Code, start from the large square in the very center of the diagram, and follow the thinnest black lines to the upper-leftmost rectangle, collecting letters from your partner's diagram as you go. There will be 8 sets of letters/numbers", "I got a graphic with different boxes and circles at the control panel", "Who has the diagram with the wiring?", "I found a diagram with lines connecting bottles", "my diagram has rectangles and circles"] #+str(x)

	spacing = random.randrange(1, 50)/10
	logstart = logstart + timedelta(seconds=spacing)
	timestamp = datetime.timestamp(logstart)

	# Add in different message types, based on game stage
	# Wrap message in log format, as in the comment above

	msgP_array.append(
		{"author": random.choice(authors),
		"message": random.choice(MsgsP),
		"timestamp": timestamp
		}
	)
#x = "Games starts"
print(msgP_array)

with open('wn_log_py.json', 'w') as json_file:
  	json.dump(msgP_array, json_file)
```

```python
#Code to generate result and pie chart
#Change msgF to msgP for different results

import re
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

Keywords = ['Who', 'Try', 'exploring', 'room', 'picture', 'found', 'diagram ', 'letters','map','invisible', 'keyboard', 'puzzle', 'press', 'solve', 'door', 'Hydraulics', 'centre', '8', 'letters', 'top', 'rectangle', 'circle', 'number', 'box', 'bottles', 'connect', 'match']

for item in msgF_array:
    temp_message = item['message'].split(' ')
    temp_message = set(temp_message)

    count = 0

    for word in Keywords:
        if word in temp_message:
            count += 1 

res = (count/len(Keywords))*100
print(res, '%')

if res == 0:
    print('Cheating has been detected')
else:
    print('No signs of detection')

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Keywords used', 'Non keywords']
sizes = [res, 100 - res]


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Cheating detection')


plt.show()
    
```
